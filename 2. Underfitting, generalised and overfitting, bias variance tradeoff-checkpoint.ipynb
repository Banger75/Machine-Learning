```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# 1. Generate non-linear data
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])  # Add noise

# 2. Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 3. Train and compare models with different complexity
degrees = [1, 4, 15]  # 1: underfit, 4: good fit, 15: overfit

plt.figure(figsize=(18, 5))

for i, d in enumerate(degrees):
    poly = PolynomialFeatures(degree=d)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    y_train_pred = model.predict(X_train_poly)
    y_test_pred = model.predict(X_test_poly)

    # Plot
    plt.subplot(1, 3, i+1)
    plt.scatter(X, y, color='gray', label='True data')
    X_line = np.linspace(0, 5, 100).reshape(-1, 1)
    X_line_poly = poly.transform(X_line)
    y_line = model.predict(X_line_poly)
    plt.plot(X_line, y_line, color='red', label=f'Model (d={d})')
    plt.title(f"Degree {d}\nTrain MSE: {mean_squared_error(y_train, y_train_pred):.2f} | Test MSE: {mean_squared_error(y_test, y_test_pred):.2f}")
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()

plt.suptitle("Underfitting vs Good Fit vs Overfitting")
plt.tight_layout()
plt.show()
```

```python
#bias variance tradeoff visualisation

import numpy as np
import matplotlib.pyplot as plt

# Simulated model complexity
model_complexity = np.linspace(1, 20, 100)

# Simulated bias (decreases), variance (increases), and total error (U-shaped)
bias_squared = np.exp(-model_complexity / 3) * 0.5
variance = np.log(model_complexity + 1) / 5
noise = 0.05
total_error = bias_squared + variance + noise

# Plot
plt.figure(figsize=(10, 6))
plt.plot(model_complexity, bias_squared, label='Bias', color='blue')
plt.plot(model_complexity, variance, label='Variance', color='red')
plt.plot(model_complexity, total_error, label='Total Error', color='green', linestyle='--')
plt.title('Bias-Variance Tradeoff')
plt.xlabel('Model Complexity')
plt.ylabel('Error')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

def true_function(x):
    return np.sin(x)

np.random.seed(42)

X_test = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)
y_true = true_function(X_test).ravel()

n_models = 100
noise_std = 0.1
degrees_to_plot = [1, 5, 15]  # low, medium, high complexity

plt.figure(figsize=(18, 5))

for i, degree in enumerate(degrees_to_plot, 1):
    all_preds = []
    for _ in range(n_models):
        X_train = np.random.uniform(0, 2 * np.pi, (30, 1))
        y_train = true_function(X_train).ravel() + np.random.normal(0, noise_std, 30)

        poly = PolynomialFeatures(degree)
        X_train_poly = poly.fit_transform(X_train)
        X_test_poly = poly.transform(X_test)

        model = LinearRegression().fit(X_train_poly, y_train)
        y_pred = model.predict(X_test_poly)
        all_preds.append(y_pred)

    all_preds = np.array(all_preds)  # shape (n_models, n_test_points)
    mean_pred = np.mean(all_preds, axis=0)

    bias_squared = (mean_pred - y_true) ** 2
    variance = np.var(all_preds, axis=0)
    total_error = bias_squared + variance + noise_std**2  # add irreducible noise

    ax = plt.subplot(1, 3, i)
    ax.plot(X_test, bias_squared, label='Bias', color='blue')
    ax.plot(X_test, variance, label='Variance', color='red')
    ax.plot(X_test, total_error, label='Total Error', color='green', linestyle='--')
    ax.set_title(f'Polynomial Degree = {degree}')
    ax.set_xlabel('x')
    ax.set_ylabel('Error')
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
